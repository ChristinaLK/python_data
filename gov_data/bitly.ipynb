```python
## write a function that processes a list of strings
## string is key for dictionary, value for key is count

def get_counts(str_seq):
    counts = {}
    for x in str_seq:
        # if x is IN the dictionary, add one
        if x in counts:
            counts[x] +=1
        # else, add count of 1 to dictionary
        else: 
            counts[x] = 1
    return counts
```

```python
counts = get_counts(time_zones)
```

```python
## want function for top time zones
## sort by values () --> put in list so it's easy to pick the top 10
def top_counts(count_dict, n=10):
    # create list of tuples. count first for sorting.  need dictionary method to iterate properly!
    pairs = [(count,tz) for tz,count in count_dict.items()]
    pairs.sort() # could you reverse sort this?
    return pairs[-n:]
```

```python
top_counts(counts)
```
